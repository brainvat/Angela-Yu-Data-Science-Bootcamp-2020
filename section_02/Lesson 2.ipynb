```python
import pandas as pd
from pandas import DataFrame as df
import matplotlib.pyplot as pp
data = pd.read_csv('movie_data_clean.csv')
```

```python
data
```

```python
X = df(data, columns = ['production_budget_usd'])
```

```python
y = df(data, columns = ['worldwide_gross_usd'])
```

```python
pp.figure(figsize=(10,6))
pp.scatter(X, y, alpha=0.3)
pp.title('Cost vs Revenue')
pp.xlabel('Production Cost')
pp.ylabel('Gross Revenue')
pp.show()
```
